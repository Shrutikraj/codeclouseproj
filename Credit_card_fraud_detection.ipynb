```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading a data set to pandas Dataframe
Credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
Credit_card_data.head()
```

```python
Credit_card_data.info()
```

```python
#checking the number of missing values in data
Credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction & fradulant transaction
Credit_card_data['Class'].value_counts()
```

```python
legit = Credit_card_data[Credit_card_data["Class"] == 0]
fraud = Credit_card_data[Credit_card_data["Class"] == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
Credit_card_data.groupby('Class').mean()
```

Under sampling

Building sample dataset containing similar distribution of fradulent and normal transaction

no.of fradulent transaction = 85

```python
legit_sample = legit.sample(n=85)
```

concatenating two dataset

```python
New_dataset = pd.concat([legit_sample, fraud] , axis=0)
#axis= 0 means adding values row vice & axis = 1 means adding values column vice
```

```python
New_dataset.head()
```

```python
New_dataset['Class'].value_counts()
```

```python
New_dataset.groupby('Class').mean()
```

```python
x = New_dataset.drop(columns='Class',axis=1)
y = New_dataset['Class']
```

```python
print(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2 , stratify = y ,random_state = 2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model =  LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data :',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on Testing data:',testing_data_accuracy)
```
